```python
import os
```

```python
def dataset(path):
 f1=open(path,'r')
 lst=[]
 for i in f1.readlines():
    if i[0]=='0':    #0
        lst.append(i)
    elif i[0]=='2':  #1
        i='1'+i[1:]
        lst.append(i)
    elif i[0]=='4':  #2
        i = '2'+i[1:]
        lst.append(i)
    elif i[0]=='5':  #3
        i = '3'+i[1:]
        lst.append(i)
    elif i[0]=='7':  #4
        i = '4'+i[1:]
        lst.append(i)
 print(lst)
 f1.close()

 f1=open(path,'w')
 f1.writelines(lst)
 f1.close()
```

```python
dataset_path='data/'
```

```python
for folder in folders:
    path=dataset_path+folder+'/labels'
    # print(path)
    for label in os.listdir(path):
        final_path=path+"/"+label
        # print(final_path)
        dataset(final_path)
```
